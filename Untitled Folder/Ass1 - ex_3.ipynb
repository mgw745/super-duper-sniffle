```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_weather(year):
    
    url = f"ftp://ftp.ncdc.noaa.gov/pub/data/ghcn/daily/by_year/{year}.csv.gz"

    df_weather = pd.read_csv(url, 
                             header=None)        

    df_weather = df_weather.iloc[:,:4] 
    
    column_names = ['station', 'datetime', 'obs_type', 'obs_value']
    df_weather.columns = column_names 
    
    df_weather['obs_value'] = df_weather['obs_value'] / 10 
    
    selection_tmax = df_weather.obs_type == 'TMAX'
    df_select = df_weather.loc[selection_tmax]
    
    df_sorted = df_select.sort_values(by=['station', 'datetime'])
    df_reset = df_sorted.reset_index(drop=True)
    df_out = df_reset.copy()
            
    return df_out
```

```python
df_weather=[]

for x in range(1870,1876):
    df_w = load_weather(x)
    df_w['year'] = x
    df_weather.append(df_w)

df_weather_period = pd.concat(df_weather, axis=0)
df_weather_period = df_weather_period.reset_index(drop=True)
df_weather_period
```

```python
df_weather_period['datetime_dt'] = pd.to_datetime(df_weather_period['datetime'].astype(str))
df_weather_period['month'] = df_weather_period['datetime_dt'].dt.month

split_vars = ['station', 'year', 'month']
apply_vars = ['obs_value']
apply_func = ['median','mean']

combined = df_weather_period.groupby(split_vars)[apply_vars].agg(apply_func)

df_weather_period['tmax_mean'] = df_weather_period.groupby(split_vars)[apply_vars].transform('mean')
df_weather_period['tmax_median'] = df_weather_period.groupby(split_vars)[apply_vars].transform('median')
df_weather_period.sort_values(by=['month', 'year'], ascending=True).reset_index(drop=True)
```

### 3.2

```python
df_wp = df_weather_period.loc[df_weather_period['station'] == 'CA006110549']
split_vars = ['month']
apply_vars = ['obs_value']

df_wp_stat = df_wp.groupby(split_vars)[apply_vars].describe() # Giver summary statistics for den månedlige temperatur 
df_wp_stat1 = df_wp_stat.drop(['count', 'std', '50%'], axis=1, level=1) # Dropper søjler på level 1

new_cols = ['mean', '25%', '75%', 'min', 'max']

df_wp_stat1 = df_wp_stat1.reindex(new_cols, axis=1, level=1)
```

```python
fig, ax = plt.subplots()

df_wp_stat1.plot(style=['r','b--','b--','bo','bo',], markersize=2, xlabel='month', ylabel='temp', title = 'Average and bands for temperature at CA006110549', ax=ax)
ax.set_aspect(0.75/ax.get_data_ratio(), adjustable='box')
handles, labels = ax.get_legend_handles_labels()
labels_new = [label.strip('()').split(',')[1] for label in labels]
plt.legend(handles, labels_new, bbox_to_anchor=(1, 1), loc='upper left')
```
