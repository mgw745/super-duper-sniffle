```python
import pandas as pd

get_url = lambda x: f'ftp://ftp.ncdc.noaa.gov/pub/data/ghcn/daily/by_year/186{x}.csv.gz'

get_dataframe = lambda x: pd.read_csv(x, compression='gzip', header=None, sep=',')

list_of_dfs = [get_dataframe(get_url(i)) for i in range(0, 4)]
   
df_weather_period = pd\
    .concat(list_of_dfs)\
    .reset_index(drop=True)

df_weather_period = df_weather_period.iloc[:, :4]
df_weather_period.columns = ['station', 'datetime', 'obs_type', 'obs_value']

df_weather_period
```
